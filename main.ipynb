```python
# ================== Imports ==================
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # Suppress TensorFlow INFO and WARNING messages
import numpy as np
import random
import tensorflow as tf
tf.get_logger().setLevel('ERROR')
from tensorflow import keras
from tensorflow.keras import layers, models, callbacks, optimizers
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import cv2
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

def build_unet(input_shape=(256, 256, 3)):
    """Build standard U-Net model"""
    inputs = layers.Input(shape=input_shape)

    # Encoder
    c1 = conv_block(inputs, 64)
    p1 = layers.MaxPooling2D((2, 2))(c1)
    p1 = layers.Dropout(0.1)(p1)

    c2 = conv_block(p1, 128)
    p2 = layers.MaxPooling2D((2, 2))(c2)
    p2 = layers.Dropout(0.1)(p2)

    c3 = conv_block(p2, 256)
    p3 = layers.MaxPooling2D((2, 2))(c3)
    p3 = layers.Dropout(0.2)(p3)

    c4 = conv_block(p3, 512)
    p4 = layers.MaxPooling2D((2, 2))(c4)
    p4 = layers.Dropout(0.2)(p4)

    # Bridge
    c5 = conv_block(p4, 1024)

    # Decoder
    u6 = layers.Conv2DTranspose(512, (2, 2), strides=(2, 2), padding='same')(c5)
    u6 = layers.concatenate([u6, c4])
    u6 = layers.Dropout(0.2)(u6)
    c6 = conv_block(u6, 512)

    u7 = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(c6)
    u7 = layers.concatenate([u7, c3])
    u7 = layers.Dropout(0.2)(u7)
    c7 = conv_block(u7, 256)

    u8 = layers.Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c7)
    u8 = layers.concatenate([u8, c2])
    u8 = layers.Dropout(0.1)(u8)
    c8 = conv_block(u8, 128)

    u9 = layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c8)
    u9 = layers.concatenate([u9, c1])
    u9 = layers.Dropout(0.1)(u9)
    c9 = conv_block(u9, 64)

    seg_output = layers.Conv2D(1, (1, 1), activation='sigmoid', name='segmentation')(c9)

    model = models.Model(inputs=[inputs], outputs=[seg_output])
    return model

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Configuration
class Config:
    SEG_BASE_PATH = '/kaggle/input/brisc2025/brisc2025/segmentation_task'
    CLS_BASE_PATH = '/kaggle/input/brisc2025/brisc2025/classification_task'
    IMG_HEIGHT = 256
    IMG_WIDTH = 256
    BATCH_SIZE = 16
    EPOCHS = 100
    LEARNING_RATE = 1e-4
    VALIDATION_SPLIT = 0.2
    PATIENCE = 15
config = Config()

# ================== Data Loading and Preprocessing ==================

class DataGenerator(tf.keras.utils.Sequence):
    def __init__(self, image_paths, mask_paths, batch_size, img_size, augment=False, class_labels=None, num_classes=2, for_task='segmentation', drop_last=True):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.batch_size = batch_size
        self.img_size = img_size
        self.augment = augment
        self.class_labels = class_labels
        self.num_classes = num_classes
        self.for_task = for_task  # 'segmentation', 'classification', or 'joint'
        self.drop_last = drop_last
        self.on_epoch_end()

    def __len__(self):
        if self.drop_last:
            return len(self.image_paths) // self.batch_size
        else:
            return int(np.ceil(len(self.image_paths) / self.batch_size))

    def __getitem__(self, index):
        indexes = self.indexes[index * self.batch_size:(index + 1) * self.batch_size]
        # Drop the last batch if it's smaller than batch_size and drop_last is True
        if self.drop_last and len(indexes) < self.batch_size:
            raise IndexError("Batch size mismatch: dropping last incomplete batch for distributed training.")
        batch_images = []
        batch_masks = []
        batch_labels = []

        for i in indexes:
            img, mask = self.load_image_mask(self.image_paths[i], self.mask_paths[i])
            if self.augment:
                img, mask = self.augment_data(img, mask)
            batch_images.append(img)
            batch_masks.append(mask)
            if self.class_labels is not None:
                label = self.class_labels[i]
                # One-hot encode
                label_onehot = np.zeros(self.num_classes)
                label_onehot[int(label)] = 1.0
                batch_labels.append(label_onehot)

        if self.for_task == 'segmentation':
            return np.array(batch_images), np.array(batch_masks)
        elif self.for_task == 'classification':
            arr_images = np.array(batch_images)
            arr_labels = np.array(batch_labels)
            return arr_images, arr_labels
        elif self.for_task == 'joint':
            return np.array(batch_images), {'segmentation': np.array(batch_masks), 'classification': np.array(batch_labels)}
        else:
            raise ValueError('Unknown for_task type')

    def on_epoch_end(self):
        self.indexes = np.arange(len(self.image_paths))
        np.random.shuffle(self.indexes)

    def load_image_mask(self, img_path, mask_path):
        # Load image
        img = cv2.imread(img_path)
        if img is None:
            img = np.zeros((self.img_size[0], self.img_size[1], 3), dtype=np.uint8)
        else:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, self.img_size)
        img = img.astype(np.float32) / 255.0

        # Load mask
        mask = None
        if mask_path is not None:
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            if mask is None:
                mask = np.zeros(self.img_size, dtype=np.uint8)
            else:
                mask = cv2.resize(mask, self.img_size)
            mask = mask.astype(np.float32) / 255.0
            mask = np.expand_dims(mask, axis=-1)
        else:
            mask = np.zeros((*self.img_size, 1), dtype=np.float32)

        return img, mask

    def augment_data(self, img, mask):
        # Random horizontal flip
        if np.random.random() > 0.5:
            img = np.fliplr(img)
            mask = np.fliplr(mask)

        # Random vertical flip
        if np.random.random() > 0.5:
            img = np.flipud(img)
            mask = np.flipud(mask)

        # Random rotation
        if np.random.random() > 0.5:
            angle = np.random.uniform(-15, 15)
            h, w = img.shape[:2]
            M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
            img = cv2.warpAffine(img, M, (w, h))
            mask = cv2.warpAffine(mask, M, (w, h))
            mask = np.expand_dims(mask, axis=-1) if len(mask.shape) == 2 else mask

        # Random brightness
        if np.random.random() > 0.5:
            factor = np.random.uniform(0.8, 1.2)
            img = np.clip(img * factor, 0, 1)

        # Random contrast
        if np.random.random() > 0.5:
            factor = np.random.uniform(0.8, 1.2)
            img = np.clip((img - 0.5) * factor + 0.5, 0, 1)

        return img, mask
# ========== Classification and Joint Model Evaluation ==========
def evaluate_classifier(model, test_gen, model_name):
    print(f"\nEvaluating {model_name}...")
    all_acc = []
    all_preds = []
    all_labels = []
    for i in range(len(test_gen)):
        x, y = test_gen[i]
        pred = model.predict(x, verbose=0)
        acc = np.mean(np.argmax(pred, axis=1) == np.argmax(y, axis=1))
        all_acc.append(acc)
        all_preds.extend(np.argmax(pred, axis=1))
        all_labels.extend(np.argmax(y, axis=1))
    print(f"{model_name} Accuracy: {np.mean(all_acc):.4f}")
    print(classification_report(all_labels, all_preds))
    return {
        'accuracy': np.mean(all_acc),
        'confusion_matrix': confusion_matrix(all_labels, all_preds).tolist(),
        'classification_report': classification_report(all_labels, all_preds, output_dict=True)
    }

def plot_classifier_history(history, model_name):
    plt.figure(figsize=(8, 4))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(f'{model_name} - Loss & Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'{model_name}_training_history.png')
    plt.show()

def evaluate_joint_model(model, test_gen, model_name):
    print(f"\nEvaluating {model_name}...")
    all_dice, all_iou, all_pixel_acc, all_acc = [], [], [], []
    all_preds, all_labels = [], []
    for i in range(len(test_gen)):
        x, y = test_gen[i]
        pred = model.predict(x, verbose=0)
        seg_pred = pred[0]
        class_pred = pred[1]
        seg_true = y['segmentation']
        class_true = y['classification']
        for j in range(len(x)):
            dice = dice_coef(seg_true[j], seg_pred[j]).numpy()
            iou = iou_metric(seg_true[j], seg_pred[j]).numpy()
            pixel_acc = pixel_accuracy(seg_true[j], seg_pred[j]).numpy()
            acc = np.argmax(class_pred[j]) == np.argmax(class_true[j])
            all_dice.append(dice)
            all_iou.append(iou)
            all_pixel_acc.append(pixel_acc)
            all_acc.append(acc)
            all_preds.append(np.argmax(class_pred[j]))
            all_labels.append(np.argmax(class_true[j]))
    print(f"{model_name} Results:")
    print(f"Mean Dice: {np.mean(all_dice):.4f}, IoU: {np.mean(all_iou):.4f}, Pixel Acc: {np.mean(all_pixel_acc):.4f}, Class Acc: {np.mean(all_acc):.4f}")
    print(classification_report(all_labels, all_preds))
    return {
        'dice': np.mean(all_dice),
        'iou': np.mean(all_iou),
        'pixel_accuracy': np.mean(all_pixel_acc),
        'classification_accuracy': np.mean(all_acc),
        'confusion_matrix': confusion_matrix(all_labels, all_preds).tolist(),
        'classification_report': classification_report(all_labels, all_preds, output_dict=True)
    }

def plot_joint_history(history, model_name):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    axes[0].plot(history.history['segmentation_loss'], label='Train Seg Loss')
    axes[0].plot(history.history['val_segmentation_loss'], label='Val Seg Loss')
    axes[0].set_title('Segmentation Loss')
    axes[0].legend()
    axes[1].plot(history.history['classification_loss'], label='Train Class Loss')
    axes[1].plot(history.history['val_classification_loss'], label='Val Class Loss')
    axes[1].set_title('Classification Loss')
    axes[1].legend()
    plt.suptitle(f'{model_name} - Loss Curves')
    plt.tight_layout()
    plt.savefig(f'{model_name}_training_history.png')
    plt.show()

def load_dataset_paths():
    """Load all image and mask paths from the dataset"""
    train_images_dir = os.path.join(config.SEG_BASE_PATH, 'train', 'images')
    train_masks_dir = os.path.join(config.SEG_BASE_PATH, 'train', 'masks')
    test_images_dir = os.path.join(config.SEG_BASE_PATH, 'test', 'images')
    test_masks_dir = os.path.join(config.SEG_BASE_PATH, 'test', 'masks')

    train_images = []
    train_masks = []
    test_images = []
    test_masks = []

    if os.path.exists(train_images_dir) and os.path.exists(train_masks_dir):
        train_images = [os.path.join(train_images_dir, fname) for fname in os.listdir(train_images_dir) if fname.lower().endswith(('.png', '.jpg', '.jpeg'))]
        train_masks = [os.path.join(train_masks_dir, fname) for fname in os.listdir(train_masks_dir) if fname.lower().endswith(('.png', '.jpg', '.jpeg'))]
        train_images.sort()
        train_masks.sort()
    if os.path.exists(test_images_dir) and os.path.exists(test_masks_dir):
        test_images = [os.path.join(test_images_dir, fname) for fname in os.listdir(test_images_dir) if fname.lower().endswith(('.png', '.jpg', '.jpeg'))]
        test_masks = [os.path.join(test_masks_dir, fname) for fname in os.listdir(test_masks_dir) if fname.lower().endswith(('.png', '.jpg', '.jpeg'))]
        test_images.sort()
        test_masks.sort()
    return train_images, train_masks, test_images, test_masks

# Classification dataset loader
def load_classification_paths():
    class_map = {'glioma': 0, 'meningioma': 1, 'no_tumor': 2, 'pituitary': 3}
    def get_paths_labels(base_dir):
        img_paths = []
        labels = []
        for cls_name, cls_idx in class_map.items():
            cls_dir = os.path.join(base_dir, cls_name)
            if os.path.exists(cls_dir):
                for fname in os.listdir(cls_dir):
                    if fname.lower().endswith(('.png', '.jpg', '.jpeg')):
                        img_paths.append(os.path.join(cls_dir, fname))
                        labels.append(cls_idx)
        return img_paths, labels
    train_dir = os.path.join(config.CLS_BASE_PATH, 'train')
    test_dir = os.path.join(config.CLS_BASE_PATH, 'test')
    train_imgs, train_labels = get_paths_labels(train_dir)
    test_imgs, test_labels = get_paths_labels(test_dir)
    return train_imgs, train_labels, test_imgs, test_labels



# ========== Model Utility Blocks ==========
def conv_block(inputs, filters, kernel_size=3, activation='relu', padding='same'):
    x = layers.Conv2D(filters, kernel_size, padding=padding)(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation(activation)(x)
    x = layers.Conv2D(filters, kernel_size, padding=padding)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation(activation)(x)
    return x

def attention_gate(x, g, filters):
    g_conv = layers.Conv2D(filters, 1, padding='same')(g)
    g_conv = layers.BatchNormalization()(g_conv)
    x_conv = layers.Conv2D(filters, 1, padding='same')(x)
    x_conv = layers.BatchNormalization()(x_conv)
    psi = layers.Activation('relu')(layers.add([g_conv, x_conv]))
    psi = layers.Conv2D(1, 1, padding='same')(psi)
    psi = layers.BatchNormalization()(psi)
    psi = layers.Activation('sigmoid')(psi)
    return layers.multiply([x, psi])

# ========== Classification Head and Joint Model ==========
def classification_head(inputs, num_classes=2, classifier_type='simple'):
    """Add a classification head on top of encoder features."""
    x = layers.GlobalAveragePooling2D()(inputs)
    if classifier_type == 'simple':
        x = layers.Dense(64, activation='relu')(x)
        x = layers.Dropout(0.5)(x)
    elif classifier_type == 'deep':
        x = layers.Dense(128, activation='relu')(x)
        x = layers.Dropout(0.5)(x)
        x = layers.Dense(64, activation='relu')(x)
        x = layers.Dropout(0.5)(x)
    class_output = layers.Dense(num_classes, activation='softmax', name='classification')(x)
    return class_output

def build_joint_unet(input_shape=(256, 256, 3), num_classes=2, classifier_type='simple'):
    """Build a U-Net with both segmentation and classification heads."""
    inputs = layers.Input(shape=input_shape)

    # Encoder (same as U-Net)
    c1 = conv_block(inputs, 64)
    p1 = layers.MaxPooling2D((2, 2))(c1)
    p1 = layers.Dropout(0.1)(p1)

    c2 = conv_block(p1, 128)
    p2 = layers.MaxPooling2D((2, 2))(c2)
    p2 = layers.Dropout(0.1)(p2)

    c3 = conv_block(p2, 256)
    p3 = layers.MaxPooling2D((2, 2))(c3)
    p3 = layers.Dropout(0.2)(p3)

    c4 = conv_block(p3, 512)
    p4 = layers.MaxPooling2D((2, 2))(c4)
    p4 = layers.Dropout(0.2)(p4)

    # Bridge
    c5 = conv_block(p4, 1024)

    # Decoder (same as U-Net)
    u6 = layers.Conv2DTranspose(512, (2, 2), strides=(2, 2), padding='same')(c5)
    u6 = layers.concatenate([u6, c4])
    u6 = layers.Dropout(0.2)(u6)
    c6 = conv_block(u6, 512)

    u7 = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(c6)
    u7 = layers.concatenate([u7, c3])
    u7 = layers.Dropout(0.2)(u7)
    c7 = conv_block(u7, 256)

    u8 = layers.Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c7)
    u8 = layers.concatenate([u8, c2])
    u8 = layers.Dropout(0.1)(u8)
    c8 = conv_block(u8, 128)

    u9 = layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c8)
    u9 = layers.concatenate([u9, c1])
    u9 = layers.Dropout(0.1)(u9)
    c9 = conv_block(u9, 64)

    seg_output = layers.Conv2D(1, (1, 1), activation='sigmoid', name='segmentation')(c9)

    # Classification head on bridge features
    class_output = classification_head(c5, num_classes=num_classes, classifier_type=classifier_type)

    model = models.Model(inputs=[inputs], outputs=[seg_output, class_output])
    return model

def build_attention_unet(input_shape=(256, 256, 3)):
    """Build Attention U-Net model"""
    inputs = layers.Input(shape=input_shape)

    # Encoder
    c1 = conv_block(inputs, 64)
    p1 = layers.MaxPooling2D((2, 2))(c1)
    p1 = layers.Dropout(0.1)(p1)

    c2 = conv_block(p1, 128)
    p2 = layers.MaxPooling2D((2, 2))(c2)
    p2 = layers.Dropout(0.1)(p2)

    c3 = conv_block(p2, 256)
    p3 = layers.MaxPooling2D((2, 2))(c3)
    p3 = layers.Dropout(0.2)(p3)

    c4 = conv_block(p3, 512)
    p4 = layers.MaxPooling2D((2, 2))(c4)
    p4 = layers.Dropout(0.2)(p4)

    # Bridge
    c5 = conv_block(p4, 1024)

    # Decoder with Attention Gates
    u6 = layers.Conv2DTranspose(512, (2, 2), strides=(2, 2), padding='same')(c5)
    a6 = attention_gate(c4, u6, 256)
    u6 = layers.concatenate([u6, a6])
    u6 = layers.Dropout(0.2)(u6)
    c6 = conv_block(u6, 512)

    u7 = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(c6)
    a7 = attention_gate(c3, u7, 128)
    u7 = layers.concatenate([u7, a7])
    u7 = layers.Dropout(0.2)(u7)
    c7 = conv_block(u7, 256)

    u8 = layers.Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c7)
    a8 = attention_gate(c2, u8, 64)
    u8 = layers.concatenate([u8, a8])
    u8 = layers.Dropout(0.1)(u8)
    c8 = conv_block(u8, 128)

    u9 = layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c8)
    a9 = attention_gate(c1, u9, 32)
    u9 = layers.concatenate([u9, a9])
    u9 = layers.Dropout(0.1)(u9)
    c9 = conv_block(u9, 64)

    outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(c9)

    model = models.Model(inputs=[inputs], outputs=[outputs])
    return model


# ================== Loss Functions and Metrics ==================

def dice_coef(y_true, y_pred, smooth=1e-6):
    """Dice coefficient for evaluation"""
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    """Dice loss for training"""
    return 1 - dice_coef(y_true, y_pred)

def iou_metric(y_true, y_pred, smooth=1e-6):
    """IoU metric"""
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    union = tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)

def pixel_accuracy(y_true, y_pred):
    """Pixel accuracy metric"""
    y_pred = tf.keras.backend.round(y_pred)
    correct = tf.keras.backend.equal(y_true, y_pred)
    return tf.keras.backend.mean(tf.keras.backend.cast(correct, tf.float32))

def combined_loss(y_true, y_pred):
    """Combined BCE + Dice loss"""
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    dice = dice_loss(y_true, y_pred)
    return bce + dice

# ---- Classification Loss and Metrics ----
def classification_loss(y_true, y_pred):
    """Categorical cross-entropy loss for classification"""
    return tf.keras.losses.categorical_crossentropy(y_true, y_pred)

def classification_accuracy(y_true, y_pred):
    return tf.keras.metrics.categorical_accuracy(y_true, y_pred)

def classification_precision(y_true, y_pred):
    return tf.keras.metrics.Precision()(y_true, y_pred)

def classification_recall(y_true, y_pred):
    return tf.keras.metrics.Recall()(y_true, y_pred)

def classification_f1(y_true, y_pred):
    precision = tf.keras.metrics.Precision()(y_true, y_pred)
    recall = tf.keras.metrics.Recall()(y_true, y_pred)
    return 2 * (precision * recall) / (precision + recall + 1e-7)

# ================== Training Functions ==================

def train_model(model, model_name, train_gen, val_gen, epochs=100):
    """Train the segmentation-only model with callbacks"""
    model.compile(
        optimizer=optimizers.Adam(learning_rate=config.LEARNING_RATE),
        loss=combined_loss,
        metrics=[dice_coef, iou_metric, pixel_accuracy]
    )
    checkpoint = callbacks.ModelCheckpoint(
        f'{model_name}_best.h5',
        monitor='val_dice_coef',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    early_stop = callbacks.EarlyStopping(
        monitor='val_dice_coef',
        mode='max',
        patience=config.PATIENCE,
        restore_best_weights=True,
        verbose=1
    )
    reduce_lr = callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    )
    steps_per_epoch = len(train_gen)
    validation_steps = len(val_gen)
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        steps_per_epoch=steps_per_epoch,
        validation_steps=validation_steps,
        epochs=epochs,
        callbacks=[checkpoint, early_stop, reduce_lr],
        verbose=1
    )
    return history

def train_joint_model(model, model_name, train_gen, val_gen, epochs=100, seg_loss=combined_loss, class_loss='categorical_crossentropy', seg_weight=1.0, class_weight=1.0):
    """Train the joint segmentation-classification model with callbacks"""
    model.compile(
        optimizer=optimizers.Adam(learning_rate=config.LEARNING_RATE),
        loss={
            'segmentation': seg_loss,
            'classification': class_loss
        },
        loss_weights={
            'segmentation': seg_weight,
            'classification': class_weight
        },
        metrics={
            'segmentation': [dice_coef, iou_metric, pixel_accuracy],
            'classification': ['accuracy']
        }
    )
    checkpoint = callbacks.ModelCheckpoint(
        f'{model_name}_best.h5',
        monitor='val_segmentation_dice_coef',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    early_stop = callbacks.EarlyStopping(
        monitor='val_segmentation_dice_coef',
        mode='max',
        patience=config.PATIENCE,
        restore_best_weights=True,
        verbose=1
    )
    reduce_lr = callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    )
    steps_per_epoch = len(train_gen)
    validation_steps = len(val_gen)
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        steps_per_epoch=steps_per_epoch,
        validation_steps=validation_steps,
        epochs=epochs,
        callbacks=[checkpoint, early_stop, reduce_lr],
        verbose=1
    )
    return history

# ================== Evaluation Function ==================

def evaluate_model(model, test_gen, model_name):
    """Evaluate model on test set"""
    print(f"\nEvaluating {model_name}...")

    all_dice = []
    all_iou = []
    all_pixel_acc = []

    for i in range(len(test_gen)):
        x, y = test_gen[i]
        pred = model.predict(x, verbose=0)

        for j in range(len(x)):
            dice = dice_coef(y[j], pred[j]).numpy()
            iou = iou_metric(y[j], pred[j]).numpy()
            pixel_acc = pixel_accuracy(y[j], pred[j]).numpy()

            all_dice.append(dice)
            all_iou.append(iou)
            all_pixel_acc.append(pixel_acc)

    print(f"{model_name} Results:")
    print(f"Mean Dice Coefficient: {np.mean(all_dice):.4f} ± {np.std(all_dice):.4f}")
    print(f"Mean IoU: {np.mean(all_iou):.4f} ± {np.std(all_iou):.4f}")
    print(f"Mean Pixel Accuracy: {np.mean(all_pixel_acc):.4f} ± {np.std(all_pixel_acc):.4f}")

    return {
        'dice': np.mean(all_dice),
        'iou': np.mean(all_iou),
        'pixel_accuracy': np.mean(all_pixel_acc)
    }

def plot_training_history(history, model_name):
    """Plot training history"""
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # Loss
    axes[0, 0].plot(history.history['loss'], label='Train Loss')
    axes[0, 0].plot(history.history['val_loss'], label='Val Loss')
    axes[0, 0].set_title(f'{model_name} - Loss')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(True)

    # Dice Coefficient
    axes[0, 1].plot(history.history['dice_coef'], label='Train Dice')
    axes[0, 1].plot(history.history['val_dice_coef'], label='Val Dice')
    axes[0, 1].set_title(f'{model_name} - Dice Coefficient')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('Dice Coefficient')
    axes[0, 1].legend()
    axes[0, 1].grid(True)

    # IoU
    axes[1, 0].plot(history.history['iou_metric'], label='Train IoU')
    axes[1, 0].plot(history.history['val_iou_metric'], label='Val IoU')
    axes[1, 0].set_title(f'{model_name} - IoU')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('IoU')
    axes[1, 0].legend()
    axes[1, 0].grid(True)

    # Pixel Accuracy
    axes[1, 1].plot(history.history['pixel_accuracy'], label='Train Pixel Acc')
    axes[1, 1].plot(history.history['val_pixel_accuracy'], label='Val Pixel Acc')
    axes[1, 1].set_title(f'{model_name} - Pixel Accuracy')
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Pixel Accuracy')
    axes[1, 1].legend()
    axes[1, 1].grid(True)

    plt.tight_layout()
    plt.savefig(f'{model_name}_training_history.png')
    plt.show()

def visualize_predictions(model, test_gen, num_samples=5):
    """Visualize model predictions"""
    fig, axes = plt.subplots(num_samples, 3, figsize=(12, 4*num_samples))

    for i in range(min(num_samples, len(test_gen))):
        x, y = test_gen[i]
        pred = model.predict(x, verbose=0)

        idx = np.random.randint(0, len(x))

        axes[i, 0].imshow(x[idx])
        axes[i, 0].set_title('Input Image')
        axes[i, 0].axis('off')

        axes[i, 1].imshow(y[idx, :, :, 0], cmap='gray')
        axes[i, 1].set_title('Ground Truth')
        axes[i, 1].axis('off')

        axes[i, 2].imshow(pred[idx, :, :, 0], cmap='gray')
        axes[i, 2].set_title('Prediction')
        axes[i, 2].axis('off')

    plt.tight_layout()
    plt.show()

# ================== Main Training Pipeline ==================


def main():
    # Enable multi-GPU training using MirroredStrategy
    print("Starting BRISC 2025 Segmentation Pipeline")
    print("="*50)

    # Load dataset paths
    train_images, train_masks, test_images, test_masks = load_dataset_paths()

    if len(train_images) == 0:
        print("No training data found. Please check the dataset path.")
        return

    # Split train into train and validation
    train_img_paths, val_img_paths, train_mask_paths, val_mask_paths = train_test_split(
        train_images, train_masks,
        test_size=config.VALIDATION_SPLIT,
        random_state=42
    )

    print(f"Training samples: {len(train_img_paths)}")
    print(f"Validation samples: {len(val_img_paths)}")
    print(f"Test samples: {len(test_images)}")


    # Load classification dataset
    cls_train_imgs, cls_train_labels, cls_test_imgs, cls_test_labels = load_classification_paths()
    # Split classification train into train/val
    cls_train_imgs, cls_val_imgs, cls_train_labels, cls_val_labels = train_test_split(
        cls_train_imgs, cls_train_labels, test_size=config.VALIDATION_SPLIT, random_state=42, stratify=cls_train_labels)

    # Segmentation data generators
    train_gen = DataGenerator(
        train_img_paths, train_mask_paths,
        config.BATCH_SIZE,
        (config.IMG_HEIGHT, config.IMG_WIDTH),
        augment=True,
        class_labels=None,
        num_classes=2,
        for_task='segmentation',
        drop_last=True
    )
    val_gen = DataGenerator(
        val_img_paths, val_mask_paths,
        config.BATCH_SIZE,
        (config.IMG_HEIGHT, config.IMG_WIDTH),
        augment=False,
        class_labels=None,
        num_classes=2,
        for_task='segmentation',
        drop_last=True
    )
    test_gen = None
    if len(test_images) > 0 and len(test_masks) > 0:
        test_gen = DataGenerator(
            test_images, test_masks,
            config.BATCH_SIZE,
            (config.IMG_HEIGHT, config.IMG_WIDTH),
            augment=False,
            class_labels=None,
            num_classes=2,
            for_task='segmentation',
            drop_last=True
        )

    # Classification data generators
    train_gen_class = DataGenerator(
        cls_train_imgs, [None]*len(cls_train_imgs),
        config.BATCH_SIZE,
        (config.IMG_HEIGHT, config.IMG_WIDTH),
        augment=True,
        class_labels=cls_train_labels,
        num_classes=4,
        for_task='classification',
        drop_last=True
    )
    val_gen_class = DataGenerator(
        cls_val_imgs, [None]*len(cls_val_imgs),
        config.BATCH_SIZE,
        (config.IMG_HEIGHT, config.IMG_WIDTH),
        augment=False,
        class_labels=cls_val_labels,
        num_classes=4,
        for_task='classification',
        drop_last=True
    )
    test_gen_class = None
    if len(cls_test_imgs) > 0:
        test_gen_class = DataGenerator(
            cls_test_imgs, [None]*len(cls_test_imgs),
            config.BATCH_SIZE,
            (config.IMG_HEIGHT, config.IMG_WIDTH),
            augment=False,
            class_labels=cls_test_labels,
            num_classes=4,
            for_task='classification',
            drop_last=True
        )

    # Joint data generators (use segmentation images/masks and real classification labels if available)
    # For demonstration, use segmentation images/masks and classification labels from the classification dataset (if available)
    # Here, we match the segmentation images to the classification labels by filename (if possible)
    # If not possible, fallback to zeros as before
    seg_train_labels = [0]*len(train_img_paths)
    seg_val_labels = [0]*len(val_img_paths)
    seg_test_labels = [0]*len(test_images)
    # Try to match classification labels to segmentation images by filename (strip extension)
    train_img_basenames = [os.path.splitext(os.path.basename(p))[0] for p in train_img_paths]
    val_img_basenames = [os.path.splitext(os.path.basename(p))[0] for p in val_img_paths]
    test_img_basenames = [os.path.splitext(os.path.basename(p))[0] for p in test_images]
    # Build a mapping from classification image basename to label
    cls_train_map = {os.path.splitext(os.path.basename(p))[0]: l for p, l in zip(cls_train_imgs, cls_train_labels)}
    cls_val_map = {os.path.splitext(os.path.basename(p))[0]: l for p, l in zip(cls_val_imgs, cls_val_labels)}
    cls_test_map = {os.path.splitext(os.path.basename(p))[0]: l for p, l in zip(cls_test_imgs, cls_test_labels)}
    # Assign labels if possible
    seg_train_labels = [cls_train_map.get(b, 0) for b in train_img_basenames]
    seg_val_labels = [cls_val_map.get(b, 0) for b in val_img_basenames]
    seg_test_labels = [cls_test_map.get(b, 0) for b in test_img_basenames]

    train_gen_joint = DataGenerator(
        train_img_paths, train_mask_paths,
        config.BATCH_SIZE,
        (config.IMG_HEIGHT, config.IMG_WIDTH),
        augment=True,
        class_labels=seg_train_labels,
        num_classes=4,
        for_task='joint',
        drop_last=True
    )
    val_gen_joint = DataGenerator(
        val_img_paths, val_mask_paths,
        config.BATCH_SIZE,
        (config.IMG_HEIGHT, config.IMG_WIDTH),
        augment=False,
        class_labels=seg_val_labels,
        num_classes=4,
        for_task='joint',
        drop_last=True
    )
    test_gen_joint = None
    if len(test_images) > 0 and len(test_masks) > 0:
        test_gen_joint = DataGenerator(
            test_images, test_masks,
            config.BATCH_SIZE,
            (config.IMG_HEIGHT, config.IMG_WIDTH),
            augment=False,
            class_labels=seg_test_labels,
            num_classes=4,
            for_task='joint',
            drop_last=True
        )

    # ========== 1. U-Net for Segmentation ==========
    print("\n" + "="*50)
    print("Training U-Net for Segmentation")
    print("="*50)
    strategy = tf.distribute.MirroredStrategy()
    print(f"Number of devices: {strategy.num_replicas_in_sync}")
    with strategy.scope():
        unet_model = build_unet(input_shape=(config.IMG_HEIGHT, config.IMG_WIDTH, 3))
        print(f"U-Net Parameters: {unet_model.count_params():,}")
        unet_history = train_model(
            unet_model,
            "unet_segmentation",
            train_gen,
            val_gen,
            epochs=config.EPOCHS
        )
    # Only save the best model (already handled by ModelCheckpoint)
    pass
    # Clear memory after training
    import gc
    tf.keras.backend.clear_session()
    gc.collect()
    print("U-Net segmentation model saved and memory cleared.")

    # ========== 2. U-Net for Classification ==========
    print("\n" + "="*50)
    print("Training U-Net for Classification")
    print("="*50)
    def build_unet_classifier(input_shape=(256, 256, 3), num_classes=2, classifier_type='simple'):
        inputs = layers.Input(shape=input_shape)
        c1 = conv_block(inputs, 64)
        p1 = layers.MaxPooling2D((2, 2))(c1)
        p1 = layers.Dropout(0.1)(p1)
        c2 = conv_block(p1, 128)
        p2 = layers.MaxPooling2D((2, 2))(c2)
        p2 = layers.Dropout(0.1)(p2)
        c3 = conv_block(p2, 256)
        p3 = layers.MaxPooling2D((2, 2))(c3)
        p3 = layers.Dropout(0.2)(p3)
        c4 = conv_block(p3, 512)
        p4 = layers.MaxPooling2D((2, 2))(c4)
        p4 = layers.Dropout(0.2)(p4)
        c5 = conv_block(p4, 1024)
        class_output = classification_head(c5, num_classes=num_classes, classifier_type=classifier_type)
        model = models.Model(inputs=[inputs], outputs=[class_output])
        return model

    # Run classification model on single device (no MirroredStrategy)
    unet_classifier = build_unet_classifier(input_shape=(config.IMG_HEIGHT, config.IMG_WIDTH, 3), num_classes=4)
    print(f"U-Net Classifier Parameters: {unet_classifier.count_params():,}")
    unet_classifier.compile(
        optimizer=optimizers.Adam(learning_rate=config.LEARNING_RATE),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    steps_per_epoch = len(train_gen_class)
    validation_steps = len(val_gen_class)
    classifier_checkpoint = callbacks.ModelCheckpoint(
        'unet_classifier_best.h5',
        monitor='val_accuracy',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    classifier_early_stop = callbacks.EarlyStopping(
        monitor='val_accuracy',
        mode='max',
        patience=config.PATIENCE,
        restore_best_weights=True,
        verbose=1
    )
    classifier_reduce_lr = callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    )
    unet_classifier_history = unet_classifier.fit(
        train_gen_class,
        validation_data=val_gen_class,
        steps_per_epoch=steps_per_epoch,
        validation_steps=validation_steps,
        epochs=config.EPOCHS,
        callbacks=[classifier_checkpoint, classifier_early_stop, classifier_reduce_lr],
        verbose=1
    )
    # Clear memory after training
    import gc
    tf.keras.backend.clear_session()
    gc.collect()
    print("U-Net classifier best model saved and memory cleared.")

    # ========== 3. Attention U-Net for Segmentation ==========
    print("\n" + "="*50)
    print("Training Attention U-Net for Segmentation")
    print("="*50)
    strategy = tf.distribute.MirroredStrategy()
    print(f"Number of devices: {strategy.num_replicas_in_sync}")
    with strategy.scope():
        attention_unet_model = build_attention_unet(input_shape=(config.IMG_HEIGHT, config.IMG_WIDTH, 3))
        print(f"Attention U-Net Parameters: {attention_unet_model.count_params():,}")
        attention_unet_history = train_model(
            attention_unet_model,
            "attention_unet_segmentation",
            train_gen,
            val_gen,
            epochs=config.EPOCHS
        )
    # Only save the best model (already handled by ModelCheckpoint)
    pass
    # Clear memory after training
    import gc
    tf.keras.backend.clear_session()
    gc.collect()
    print("Attention U-Net segmentation model saved and memory cleared.")

    # ========== 4. Joint U-Net for Segmentation + Classification ==========
    print("\n" + "="*50)
    print("Training Joint U-Net for Segmentation + Classification")
    print("="*50)
    strategy = tf.distribute.MirroredStrategy()
    print(f"Number of devices: {strategy.num_replicas_in_sync}")
    with strategy.scope():
        joint_model = build_joint_unet(input_shape=(config.IMG_HEIGHT, config.IMG_WIDTH, 3), num_classes=4, classifier_type='simple')
        print(f"Joint U-Net Parameters: {joint_model.count_params():,}")
        joint_history = train_joint_model(joint_model, "joint_unet", train_gen_joint, val_gen_joint, epochs=config.EPOCHS)
    # Only save the best model (already handled by ModelCheckpoint)
    pass
    # Clear memory after training
    import gc
    tf.keras.backend.clear_session()
    gc.collect()
    print("Joint U-Net model saved and memory cleared.")

    # ========== 5. Evaluation and Plotting ==========
    import json
    import pandas as pd
    from tensorflow.keras.models import load_model
    metrics_dict = {}


    if test_gen:
        print("\nEvaluating U-Net Segmentation...")
        unet_model = load_model('/kaggle/working/unet_segmentation_best.h5', compile=False)
        unet_results = evaluate_model(unet_model, test_gen, "U-Net")
        plot_training_history(unet_history, "U-Net")
        visualize_predictions(unet_model, test_gen, num_samples=3)
        metrics_dict['unet_segmentation'] = unet_results

        print("\nEvaluating Attention U-Net Segmentation...")
        attention_unet_model = load_model('/kaggle/working/attention_unet_segmentation_best.h5', compile=False)
        attention_unet_results = evaluate_model(attention_unet_model, test_gen, "Attention U-Net")
        plot_training_history(attention_unet_history, "Attention U-Net")
        visualize_predictions(attention_unet_model, test_gen, num_samples=3)
        metrics_dict['attention_unet_segmentation'] = attention_unet_results


    print("\nEvaluating U-Net Classification...")
    unet_classifier = load_model('/kaggle/working/unet_classifier_best.h5', compile=False)
    if test_gen_class:
        unet_classifier_results = evaluate_classifier(unet_classifier, test_gen_class, "U-Net Classifier")
        if 'history' in locals() and unet_classifier_history is not None:
            plot_classifier_history(unet_classifier_history, "U-Net Classifier")
        metrics_dict['unet_classifier'] = unet_classifier_results


    print("\nEvaluating Joint U-Net...")
    joint_model = load_model('/kaggle/working/joint_unet_best.h5', compile=False)
    if test_gen_joint:
        joint_unet_results = evaluate_joint_model(joint_model, test_gen_joint, "Joint U-Net")
        if 'joint_history' in locals() and joint_history is not None:
            plot_joint_history(joint_history, "Joint U-Net")
        metrics_dict['joint_unet'] = joint_unet_results

        # ========== 6. Save Metrics and Comparison Plots ==========
    def convert(o):
        import numpy as np
        if isinstance(o, (np.integer, np.int_)):
            return int(o)
        if isinstance(o, (np.floating, np.float_)):
            return float(o)
        if isinstance(o, np.ndarray):
            return o.tolist()
        return str(o)

    with open('/kaggle/working/model_metrics.json', 'w') as f:
        json.dump(metrics_dict, f, indent=2, default=convert)
    print("All evaluation metrics saved to /kaggle/working/model_metrics.json")

    # ========== 7. Summary Comparison Plot and Table ==========
    df = pd.DataFrame(metrics_dict).T
    df.to_csv('/kaggle/working/model_metrics_summary.csv')
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for metric in numeric_cols:
        plt.figure(figsize=(8,4))
        df[metric].plot(kind='bar')
        plt.title(f'Model Comparison: {metric}')
        plt.ylabel(metric)
        plt.xlabel('Model')
        plt.tight_layout()
        plt.savefig(f'/kaggle/working/model_comparison_{metric}.png')
        plt.close()
    print('Summary comparison plots and table saved in Kaggle working directory.')

    print("\nTraining Complete! Best models saved as 'unet_segmentation_best.h5', 'attention_unet_segmentation_best.h5', 'unet_classifier_best.h5', and 'joint_unet_best.h5'.")
    return unet_model, attention_unet_model

# Run the pipeline
if __name__ == "__main__":
    unet, attention_unet = main()
```

```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# --- CONFIG ---
IMG_DIR = '/kaggle/input/brisc2025/brisc2025/segmentation_task/test/images'
MASK_DIR = '/kaggle/input/brisc2025/brisc2025/segmentation_task/test/masks'
CLS_TEST_DIR = '/kaggle/input/brisc2025/brisc2025/classification_task/test'
IMG_SIZE = (256, 256)

# Model paths
UNET_PATH = '/kaggle/working/unet_segmentation_best.h5'
ATTN_UNET_PATH = '/kaggle/working/attention_unet_segmentation_best.h5'
UNET_CLS_PATH = '/kaggle/working/unet_classifier_best.h5'
JOINT_UNET_PATH = '/kaggle/working/joint_unet_best.h5'

# --- PICK RANDOM IMAGE ---
img_files = [f for f in os.listdir(IMG_DIR) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff'))]
if not img_files:
    raise FileNotFoundError("No images found in test images directory.")
IMG_NAME = random.choice(img_files)
IMG_PATH = os.path.join(IMG_DIR, IMG_NAME)
MASK_PATH = os.path.join(MASK_DIR, os.path.splitext(IMG_NAME)[0] + '.png')

# --- Infer true class from classification test folders ---
true_class_label = None
for class_folder in os.listdir(CLS_TEST_DIR):
    class_folder_path = os.path.join(CLS_TEST_DIR, class_folder)
    if not os.path.isdir(class_folder_path):
        continue
    if IMG_NAME in os.listdir(class_folder_path):
        true_class_label = class_folder
        break
if true_class_label is None:
    true_class_label = 'Unknown'

# --- LOAD MODELS ---
unet = load_model(UNET_PATH, compile=False)
attn_unet = load_model(ATTN_UNET_PATH, compile=False)
unet_cls = load_model(UNET_CLS_PATH, compile=False)
joint_unet = load_model(JOINT_UNET_PATH, compile=False)

# --- LOAD IMAGE & MASK ---
orig_img = cv2.imread(IMG_PATH)
if orig_img is None:
    raise FileNotFoundError(f"Image not found or cannot be read: {IMG_PATH}")
orig_img = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)
orig_mask = cv2.imread(MASK_PATH, cv2.IMREAD_GRAYSCALE)
if orig_mask is None:
    raise FileNotFoundError(f"Mask not found or cannot be read: {MASK_PATH}")

# --- PROCESS IMAGE ---
proc_img = cv2.resize(orig_img, IMG_SIZE)
proc_img = proc_img.astype(np.float32) / 255.0
input_img = np.expand_dims(proc_img, axis=0)

# --- PREDICT MASKS ---
unet_pred = unet.predict(input_img)[0, :, :, 0]
attn_unet_pred = attn_unet.predict(input_img)[0, :, :, 0]
joint_pred_mask, joint_pred_class = joint_unet.predict(input_img)
joint_pred_mask = joint_pred_mask[0, :, :, 0]
unet_cls_pred = unet_cls.predict(input_img)[0]

# --- Binarize masks ---
def binarize(mask):
    return (mask > 0.5).astype(np.float32)

unet_pred_bin = binarize(unet_pred)
attn_unet_pred_bin = binarize(attn_unet_pred)
joint_pred_mask_bin = binarize(joint_pred_mask)

# --- Overlays ---
def overlay_mask(image, mask, color=(255,0,0), alpha=0.4):
    overlay = image.copy()
    mask_rgb = np.zeros_like(image)
    mask_rgb[:,:,0] = mask * color[0]
    mask_rgb[:,:,1] = mask * color[1]
    mask_rgb[:,:,2] = mask * color[2]
    cv2.addWeighted(mask_rgb.astype(np.uint8), alpha, overlay, 1-alpha, 0, overlay)
    return overlay

orig_mask_resized = cv2.resize(orig_mask, orig_img.shape[:2][::-1])
unet_pred_resized = cv2.resize((unet_pred_bin*255).astype(np.uint8), orig_img.shape[:2][::-1])
attn_unet_pred_resized = cv2.resize((attn_unet_pred_bin*255).astype(np.uint8), orig_img.shape[:2][::-1])
joint_pred_mask_resized = cv2.resize((joint_pred_mask_bin*255).astype(np.uint8), orig_img.shape[:2][::-1])

unet_overlay = overlay_mask(orig_img, (unet_pred_resized > 127).astype(np.uint8))
attn_unet_overlay = overlay_mask(orig_img, (attn_unet_pred_resized > 127).astype(np.uint8))
joint_overlay = overlay_mask(orig_img, (joint_pred_mask_resized > 127).astype(np.uint8))

# --- Predicted classes ---
class_names = ['glioma', 'meningioma', 'no_tumor', 'pituitary']  # adjust as per your dataset
unet_cls_label = class_names[np.argmax(unet_cls_pred)]
joint_cls_label = class_names[np.argmax(joint_pred_class[0])]

# --- PLOT: All images in a row, class results as text below ---
fig, axes = plt.subplots(1, 5, figsize=(25, 6))

axes[0].imshow(orig_img)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(orig_mask, cmap='gray')
axes[1].set_title('Original Mask')
axes[1].axis('off')

axes[2].imshow(unet_overlay)
axes[2].set_title('U-Net Mask Overlay')
axes[2].axis('off')

axes[3].imshow(attn_unet_overlay)
axes[3].set_title('Attention U-Net Mask Overlay')
axes[3].axis('off')

axes[4].imshow(joint_overlay)
axes[4].set_title('Joint U-Net Mask Overlay')
axes[4].axis('off')

plt.tight_layout()
plt.show()

# --- Show class predictions as text ---
print(f'U-Net Classifier: Predicted = {unet_cls_label}, True = {true_class_label}')
print(f'Joint U-Net Class: Predicted = {joint_cls_label}, True = {true_class_label}')
```
